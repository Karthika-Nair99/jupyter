```python
import pandas as pd
import numpy as np
from urllib.request import urlopen,Request
from bs4 import BeautifulSoup
```

```python
url = "https://www.holidify.com/collections/tourist-places-in-india/"
agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36\
(KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'

request = Request(url, headers={'User-Agent': agent})

html = urlopen(request).read().decode()

```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
for a in soup.findAll(attrs={'class':'row no-gutters d-flex'}):
    rows = a.find_all('h3')
    
print(rows)
```

```python
str_cells = str(rows)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
list_rows.clear()
```

```python
list_rows = []
for row in rows:
    str_cells = str(row)
    cleantext = BeautifulSoup(str_cells, "lxml").get_text()
    list_rows.append(cleantext)
```

```python
print(list_rows)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split('.', expand=True)
df1.head(10)
```

```python
df1.pop(2)
df1.pop(0)
```

```python
df2 = df1[1].str.split(',',expand=True)
df2.head(10)
```

```python
df4=df3[0].str.split(',',expand=True)
df4.head(10)
```

```python
frames=[df2[0],df4[1],df3[1]]
```
